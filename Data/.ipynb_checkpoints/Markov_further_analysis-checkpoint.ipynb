```python
import pandas as pd
df = pd.read_csv('../data/test_markov.csv')
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# Unique movements of customers
df1 = pd.DataFrame(df.groupby('customer_ID')['location'].count())
print("Unique movement of customers :\n", df1.location.unique()) #unique behavioral movement of customers
print('\n')
print("Customers with maximum movements is/are :\n", df1[df1['location']==df1['location'].max()]) #max unique movement
print('\n')
print("Customers with minimum movements is/are :\n", df1[df1['location']==df1['location'].min()]) #min unique movement
```

```python
df2 = pd.DataFrame(df.groupby(['customer_ID','time'])['location'].sum())
df2.head(30)
```

```python
df2 = pd.DataFrame(df.groupby('customer_ID')['location'].unique())
df2.head(5)
```

```python
# Transform list of list to string only so you can split
dx = df2.explode('location') #moves next element to next line with same index
dx['location'] = dx['location'].astype(str) #ensures or converts to string
dx=dx.groupby(level=0).agg(' '.join) #joins by same index
dx1=dx.copy()
dx.head(2)
```

```python
# Method 1 keeps initial columns
dx[['loc1','loc2','loc3','loc4','loc5']] = pd.DataFrame(dx.location.str.split(' ',4, expand=True))
                        
```

```python
dx.head()
```

```python
# Method 2 drops initial columns
dfx = pd.DataFrame(dx1.location.str.split(' ',4).tolist(),
                                 columns = ['loc1','loc2','loc3','loc4','loc5'])
```
